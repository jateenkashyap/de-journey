```python
data = {
    "name":    ["Alice", "Bob", "Carol", "Dave"],
    "revenue": [1200, 800, 1500, 600],
    "region":  ["North", "South", "North", "South"]
}

df = pd.DataFrame(data)
df
```

```python
print("Type:", type(df))
print("Shape:", df.shape)
print("Dtypes:\n", df.dtypes)
print("Index:", df.index)
```

```python
# Sort to make the difference visible
print(df)
print("\n")
df_sorted = df.sort_values("revenue", ascending=False)

print("Sorted DataFrame:")
print(df_sorted)
print("\nilocl[0] - by position:", df_sorted.iloc[0]["name"])
print("loc[0] - by label:", df_sorted.loc[0]["name"])
```

```python
df_sorted = df.sort_values("revenue", ascending=False).reset_index(drop=True)

print(df_sorted)
print("\nAfter reset - iloc[0]:", df_sorted.iloc[0]["name"])
print("After reset - loc[0]:", df_sorted.loc[0]["name"])
```

```python
# Boolean mask
print(df)
print('\n')
mask = df["revenue"] > 1000
print("Mask:")
print(mask)

print("\nFiltered DataFrame:")
print(df[mask])

# Combined conditions
print("\nRevenue > 1000 AND region == North:")
print(df[(df["revenue"] > 1000) & (df["region"] == "North")])

# Using .query()
print("\nSame with .query():")
print(df.query("revenue > 1000 and region == 'North'"))

# Filter + select specific columns
print("\nFilter + select columns with .loc:")
print(df.loc[df["revenue"] > 1000, ["name", "revenue"]])
```

```python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Carol", "Dave"]
})

orders = pd.DataFrame({
    "order_id":    [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 2, 3, 5],  # notice: 4 is missing, 5 doesn't exist
    "amount":      [250, 150, 300, 175, 200]
})

print("Customers:")
print(customers)
print("\nOrders:")
print(orders)
```

```python
orders
```

```python
# INNER JOIN - only matching rows on both sides
inner = pd.merge(customers, orders, on="customer_id", how="inner")
print("INNER JOIN:")
print(inner)

# LEFT JOIN - all customers, even if no orders
left = pd.merge(customers, orders, on="customer_id", how="left")
print("\nLEFT JOIN:")
print(left)

# RIGHT JOIN - all orders, even if customer doesn't exist
right = pd.merge(customers, orders, on="customer_id", how="right")
print("\nRIGHT JOIN:")
print(right)

# OUTER JOIN - everything from both sides
outer = pd.merge(customers, orders, on="customer_id", how="outer")
print("\nOUTER JOIN:")
print(outer)
```

```python
wide_df = pd.DataFrame({
    "name":      ["Alice", "Bob"],
    "jan_sales": [1200, 800],
    "feb_sales": [1500, 900],
    "mar_sales": [1100, 750]
})

print("Wide format:")
print(wide_df)

# melt - convert wide to long
long_df = wide_df.melt(
    id_vars="name",        # column to keep as identifier
    var_name="month",      # name for the new column headers column
    value_name="sales"     # name for the values column
)

print("\nLong format:")
print(long_df)
```

```python
# Using the long_df we just created
print("Long format (input):")
print(long_df)

# pivot_table - convert long to wide
pivot_df = long_df.pivot_table(
    index="name",      # what becomes the row labels
    columns="month",   # what becomes the column headers
    values="sales"     # what fills the cells
)

print("\nWide format (output):")
print(pivot_df)
```

```python
# Realistic employee performance data
# Multiple metrics per person per month
wide_data = pd.DataFrame({
    "name":           ["Alice", "Bob", "Carol", "Dave", "Eve"],
    "region":         ["North", "South", "North", "South", "North"],
    "department":     ["Sales", "Sales", "Marketing", "Marketing", "Sales"],
    "jan_sales":      [1200, 800, 1500, 600, 900],
    "feb_sales":      [1500, 900, 1700, 750, 1100],
    "mar_sales":      [1100, 750, 1300, 500, 950],
    "jan_calls":      [45, 30, 55, 20, 35],
    "feb_calls":      [50, 35, 60, 25, 40],
    "mar_calls":      [48, 32, 58, 22, 38],
    "jan_satisfaction":[4.2, 3.8, 4.5, 3.5, 4.0],
    "feb_satisfaction":[4.4, 3.9, 4.6, 3.7, 4.1],
    "mar_satisfaction":[4.3, 4.0, 4.7, 3.6, 4.2],
})

print("Wide format:")
print(wide_data)
print(f"\nShape: {wide_data.shape}")
```

```python
# You can choose WHICH columns to melt
# id_vars = columns to keep as-is
long_sales = wide_data.melt(
    id_vars=["name", "region", "department"],  # keep these fixed
    value_vars=["jan_sales", "feb_sales", "mar_sales"],  # only melt these
    var_name="month",
    value_name="sales"
)

print("Long format - sales only:")
print(long_sales)
print(f"\nShape: {long_sales.shape}")
```

```python
# Pivot sales back to wide
pivot_back = long_sales.pivot_table(
    index=["name", "region", "department"],  # multiple index columns
    columns="month",
    values="sales"
)

print("Pivoted back:")
print(pivot_back)
```

```python
import numpy as np

df_missing = pd.DataFrame({
    "name":     ["Alice", "Bob", "Carol", "Dave", "Eve"],
    "region":   ["North", None, "North", "South", None],
    "revenue":  [1200, 800, None, 600, 900],
    "category": ["A", "B", "A", None, "B"]
})

print(df_missing)
print("\nMissing values per column:")
print(df_missing.isna().sum())
print(f"\nTotal missing: {df_missing.isna().sum().sum()}")
```

```python
df_filled = df_missing.copy()  # always work on a copy, never modify raw data

# Fill revenue with median - meaningful for numeric columns
df_filled["revenue"] = df_filled["revenue"].fillna(df_filled["revenue"].median())

# Fill region with mode (most common value)
df_filled["region"] = df_filled["region"].fillna(df_filled["region"].mode()[0])

# Fill category with a literal string
df_filled["category"] = df_filled["category"].fillna("Unknown")

print("After filling:")
print(df_filled)
print("\nMissing values remaining:")
print(df_filled.isna().sum())
```

```python
# Drop rows where ANY column is null
print("Drop any null:")
print(df_missing.dropna())

# Drop rows where ALL columns are null
print("\nDrop only if ALL null:")
print(df_missing.dropna(how="all"))

# Drop rows where specific column is null
# Use this when that column is critical to your pipeline
print("\nDrop only where revenue is null:")
print(df_missing.dropna(subset=["revenue"]))
```

```python
df3 = pd.DataFrame({
    "name":     ["Alice", "Bob", "Carol", "Dave"],
    "revenue":  [1200, 800, 1500, 600],
    "region":   ["North", "South", "North", "South"]
})

# WRONG — chained indexing, may silently fail
df3[df3["region"] == "North"]["revenue"] = 9999
print("After chained assignment (WRONG way):")
print(df3)  # revenue did NOT change — silent failure

# RIGHT — always use .loc for assignments
df3.loc[df3["region"] == "North", "revenue"] = 9999
print("\nAfter .loc assignment (RIGHT way):")
print(df3)  # revenue changed correctly
```

```python
import pandas as pd

# Realistic sales rep data
df_sales = pd.DataFrame({
    "name":       ["Alice", "Bob", "Carol", "Dave", "Eve"],
    "revenue":    [1200, 800, 1500, 600, 900],
    "region":     ["North", "South", "North", "South", "North"],
    "experience": [5, 2, 8, 1, 3],   # years of experience
    "rating":     [4.2, 3.8, 4.7, 3.2, 4.0]  # performance rating
})

df_sales
```

```python
# Categorize sales reps by revenue performance
def performance_band(revenue):
    if revenue >= 1200:
        return "Top Performer"
    elif revenue >= 800:
        return "On Track"
    else:
        return "Needs Improvement"

df_sales["performance"] = df_sales["revenue"].apply(performance_band)
df_sales
```

```python
# Lambda - when the logic is simple enough for one line
df_sales["revenue_in_thousands"] = df_sales["revenue"].apply(lambda x: round(x / 1000, 1))

# Bonus calculation - 10% for top performers, 5% for rest
df_sales["bonus"] = df_sales["revenue"].apply(lambda x: x * 0.10 if x >= 1200 else x * 0.05)

df_sales
```

```python
# Sometimes you need values from MORE than one column
# axis=1 means apply row by row

def eligibility(row):
    # Eligible for promotion if: revenue > 1000 AND experience > 3 AND rating > 4.0
    if row["revenue"] > 1000 and row["experience"] > 3 and row["rating"] > 4.0:
        return "Eligible"
    else:
        return "Not Eligible"

df_sales["promotion"] = df_sales.apply(eligibility, axis=1)
df_sales
```

```python
# map() is simpler than apply() - just replaces values using a dictionary
# Perfect for encoding or renaming categories

region_map = {
    "North": "Northern Region",
    "South": "Southern Region"
}

df_sales["region_full"] = df_sales["region"].map(region_map)
df_sales
```

```python
df_compare = pd.DataFrame({
    "name":   ["Alice", "Bob", "Carol", "Dave"],
    "region": ["North", "South", "North", "South"],
    "revenue":[1200, 800, 1500, 600]
})

# -----------------------------------------------
# map() — ONLY for replacing/looking up values
# Works on a Series, uses a dictionary or function
# Simple 1-to-1 replacement
# -----------------------------------------------

# USE map() when: you have a known lookup table
region_codes = {"North": "N", "South": "S"}
df_compare["region_code"] = df_compare["region"].map(region_codes)
# "North" → "N", "South" → "S"
# No logic, no conditions, just lookup

print("After map():")
print(df_compare)

# -----------------------------------------------
# apply() — for transformations with logic
# Works on a Series OR whole row (axis=1)
# Can have conditions, calculations, if/else
# -----------------------------------------------

# USE apply() when: you need logic or calculations
df_compare["revenue_band"] = df_compare["revenue"].apply(lambda x: "High" if x >= 1000 else "Low")
# Has a condition — map() can't do this cleanly

print("\nAfter apply():")
print(df_compare)

# -----------------------------------------------
# What happens when map() meets unknown values?
# -----------------------------------------------
region_codes_incomplete = {"North": "N"}  # South is missing!
df_compare["region_code_v2"] = df_compare["region"].map(region_codes_incomplete)

print("\nmap() with missing keys — becomes NaN:")
print(df_compare[["region", "region_code_v2"]])
# South → NaN because it wasn't in the dictionary
# apply() would never do this — it runs the function regardless
```
