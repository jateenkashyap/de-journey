```python
import pandas as pd

print("Pandas version:", pd.__version__)
```

```python
data = {
    "name":    ["Alice", "Bob", "Carol", "Dave"],
    "revenue": [1200, 800, 1500, 600],
    "region":  ["North", "South", "North", "South"]
}

df = pd.DataFrame(data)
df
```

```python
print("Type:", type(df))
print("Shape:", df.shape)
print("Dtypes:\n", df.dtypes)
print("Index:", df.index)
```

```python
# Sort to make the difference visible
print(df)
print("\n")
df_sorted = df.sort_values("revenue", ascending=False)

print("Sorted DataFrame:")
print(df_sorted)
print("\nilocl[0] - by position:", df_sorted.iloc[0]["name"])
print("loc[0] - by label:", df_sorted.loc[0]["name"])
```

```python
df_sorted = df.sort_values("revenue", ascending=False).reset_index(drop=True)

print(df_sorted)
print("\nAfter reset - iloc[0]:", df_sorted.iloc[0]["name"])
print("After reset - loc[0]:", df_sorted.loc[0]["name"])
```

```python
# Boolean mask
print(df)
print('\n')
mask = df["revenue"] > 1000
print("Mask:")
print(mask)

print("\nFiltered DataFrame:")
print(df[mask])

# Combined conditions
print("\nRevenue > 1000 AND region == North:")
print(df[(df["revenue"] > 1000) & (df["region"] == "North")])

# Using .query()
print("\nSame with .query():")
print(df.query("revenue > 1000 and region == 'North'"))

# Filter + select specific columns
print("\nFilter + select columns with .loc:")
print(df.loc[df["revenue"] > 1000, ["name", "revenue"]])
```

```python
data2 = {
    "name":     ["Alice", "Bob", "Carol", "Dave", "Eve"],
    "region":   ["North", "South", "North", "South", "North"],
    "revenue":  [1200, 800, 1500, 600, 900],
    "category": ["A", "B", "A", "B", "B"]
}

df2 = pd.DataFrame(data2)
df2
```

```python
# Total revenue per region
print("Total revenue per region:")
print(df2.groupby("region")["revenue"].sum())
```

```python
# sum, mean, count in one shot
print("Multiple aggregations:")
print(df2.groupby("region")["revenue"].agg(["sum", "mean", "count"]))
```

```python
# Group by region AND category
print("Revenue by region and category:")
print(df2.groupby(["region", "category"])["revenue"].sum())
```

```python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Carol", "Dave"]
})

orders = pd.DataFrame({
    "order_id":    [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 2, 3, 5],  # notice: 4 is missing, 5 doesn't exist
    "amount":      [250, 150, 300, 175, 200]
})

print("Customers:")
print(customers)
print("\nOrders:")
print(orders)
```

```python
# INNER JOIN - only matching rows on both sides
inner = pd.merge(customers, orders, on="customer_id", how="inner")
print("INNER JOIN:")
print(inner)

# LEFT JOIN - all customers, even if no orders
left = pd.merge(customers, orders, on="customer_id", how="left")
print("\nLEFT JOIN:")
print(left)

# RIGHT JOIN - all orders, even if customer doesn't exist
right = pd.merge(customers, orders, on="customer_id", how="right")
print("\nRIGHT JOIN:")
print(right)

# OUTER JOIN - everything from both sides
outer = pd.merge(customers, orders, on="customer_id", how="outer")
print("\nOUTER JOIN:")
print(outer)
```

```python
wide_df = pd.DataFrame({
    "name":      ["Alice", "Bob"],
    "jan_sales": [1200, 800],
    "feb_sales": [1500, 900],
    "mar_sales": [1100, 750]
})

print("Wide format:")
print(wide_df)

# melt - convert wide to long
long_df = wide_df.melt(
    id_vars="name",        # column to keep as identifier
    var_name="month",      # name for the new column headers column
    value_name="sales"     # name for the values column
)

print("\nLong format:")
print(long_df)
```

```python
# Using the long_df we just created
print("Long format (input):")
print(long_df)

# pivot_table - convert long to wide
pivot_df = long_df.pivot_table(
    index="name",      # what becomes the row labels
    columns="month",   # what becomes the column headers
    values="sales"     # what fills the cells
)

print("\nWide format (output):")
print(pivot_df)
```

```python
# Realistic employee performance data
# Multiple metrics per person per month
wide_data = pd.DataFrame({
    "name":           ["Alice", "Bob", "Carol", "Dave", "Eve"],
    "region":         ["North", "South", "North", "South", "North"],
    "department":     ["Sales", "Sales", "Marketing", "Marketing", "Sales"],
    "jan_sales":      [1200, 800, 1500, 600, 900],
    "feb_sales":      [1500, 900, 1700, 750, 1100],
    "mar_sales":      [1100, 750, 1300, 500, 950],
    "jan_calls":      [45, 30, 55, 20, 35],
    "feb_calls":      [50, 35, 60, 25, 40],
    "mar_calls":      [48, 32, 58, 22, 38],
    "jan_satisfaction":[4.2, 3.8, 4.5, 3.5, 4.0],
    "feb_satisfaction":[4.4, 3.9, 4.6, 3.7, 4.1],
    "mar_satisfaction":[4.3, 4.0, 4.7, 3.6, 4.2],
})

print("Wide format:")
print(wide_data)
print(f"\nShape: {wide_data.shape}")
```

```python
# You can choose WHICH columns to melt
# id_vars = columns to keep as-is
long_sales = wide_data.melt(
    id_vars=["name", "region", "department"],  # keep these fixed
    value_vars=["jan_sales", "feb_sales", "mar_sales"],  # only melt these
    var_name="month",
    value_name="sales"
)

print("Long format - sales only:")
print(long_sales)
print(f"\nShape: {long_sales.shape}")
```

```python
# Pivot sales back to wide
pivot_back = long_sales.pivot_table(
    index=["name", "region", "department"],  # multiple index columns
    columns="month",
    values="sales"
)

print("Pivoted back:")
print(pivot_back)
```
